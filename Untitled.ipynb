```r
library("ggplot2")
library("dplyr")
```

```r
data =read.csv("data.csv")
```

```r
data_clean = unique(data)
```

```r
#data0 = data[data$out == 0]
datos0 = data_clean %>% filter(out == 0)
datos1 = data_clean %>% filter(out == 1)
datos0 = datos0[c(1:500),]
data_clean= rbind(datos0,datos1)
```

```r
set.seed(42)
rows <- sample(nrow(data_clean))
data_clean <- data_clean[rows,]
data_clean
```

```r
write.csv(data_clean,"data_clean.csv", row.names = FALSE)
```

```r
ggplot(data_clean, aes(Angle, Power, colour = out)) + 
  geom_point()
```
